```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('FinalDataset.csv')

# Define the input and output variables
input_vars = ['MONSEX', 'NEWRACE', 'AGE', 'NEWCIT', 'HISPORIG', 'EDUCATN', 'NUMDEPEN',
              'CRIMHIST', 'CITWHERE', 'CIRC', 'COUNTY',
              'OFFTYPE2', 'OFFTYPSB', 'NWSTAT1', 'NWSTAT2', 'NWSTAT3', 'NWSTAT4', 'NWSTAT5',
              'ACCAP', 'WEAPSOC', 'NOCOUNTS', 'DISPOSIT']

output_vars = ['SENTTOT', 'SENSPLT0', 'SENTPTS', 'BOOKER2', 'BOOKERCD', 'SENTIMP',
               'AMTFINEC', 'AMTREST', 'AMTTOTAL',
               'ALTDUM', 'ALTMO', 'TIMSERVD', 'SUPRDUM', 'SUPREL']

# Create a subplot grid
num_inputs = len(input_vars)
num_outputs = len(output_vars)
fig, axes = plt.subplots(num_inputs, num_outputs, figsize=(num_outputs*4, num_inputs*4))

# Iterate over each input and output variable
for i, input_var in enumerate(input_vars):
    for j, output_var in enumerate(output_vars):
        ax = axes[i, j]
        
        # Plot the relationship based on variable types
        if data[input_var].dtype == 'object' and data[output_var].dtype == 'object':
            # Categorical input and output
            pd.crosstab(data[input_var], data[output_var]).plot(kind='bar', ax=ax)
        elif data[input_var].dtype == 'object':
            # Categorical input and numerical output
            data.groupby(input_var)[output_var].mean().plot(kind='bar', ax=ax)
        elif data[output_var].dtype == 'object':
            # Numerical input and categorical output
            for category in data[output_var].unique():
                data[data[output_var] == category].plot(kind='scatter', x=input_var, y=output_var, ax=ax, label=category)
            ax.legend()
        else:
            # Numerical input and output
            data.plot(kind='scatter', x=input_var, y=output_var, ax=ax)
        
        ax.set_xlabel(input_var)
        ax.set_ylabel(output_var)
        ax.set_title(f'{input_var} vs {output_var}')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Create a folder to store the plots
os.makedirs('plots', exist_ok=True)

# Load the dataset
data = pd.read_csv('FinalDataset.csv')

# Define the input and output variables with full descriptions
input_vars = {
    'MONSEX': 'Defendant\'s Gender',
    'NEWRACE': 'Defendant\'s Race',
    'AGE': 'Defendant\'s Age',
    'NEWCIT': 'Defendant\'s Citizenship',
    'HISPORIG': 'Defendant\'s Hispanic Origin',
    'EDUCATN': 'Defendant\'s Education Level',
    'NUMDEPEN': 'Number of Dependents',
    'CRIMHIST': 'Defendant\'s Criminal History',
    'CITWHERE': 'Defendant\'s Country of Citizenship',
    'CIRC': 'Circuit where Defendant was Sentenced',
    'COUNTY': 'County where Offense was Committed',
    'OFFTYPE2': 'Type of Offense (Version 2)',
    'OFFTYPSB': 'Type of Offense (Sourcebook Version)',
    'NWSTAT1': 'Statute 1',
    'NWSTAT2': 'Statute 2',
    'NWSTAT3': 'Statute 3',
    'NWSTAT4': 'Statute 4',
    'NWSTAT5': 'Statute 5',
    'ACCAP': 'Armed Career Criminal Status',
    'WEAPSOC': 'Weapon Enhancement Applied',
    'NOCOUNTS': 'Number of Counts of Conviction',
    'DISPOSIT': 'Disposition of the Case'
}

output_vars = {
    'SENTTOT': 'Total Prison Sentence (Months)',
    'SENSPLT0': 'Total Prison Sentence with Alternatives (Months)',
    'SENTPTS': 'Criminal History Points Applied',
    'BOOKER2': 'Post-Booker Sentencing (Version 2)',
    'BOOKERCD': 'Post-Booker Sentencing (Detailed)',
    'SENTIMP': 'Type of Sentence Imposed',
    'AMTFINEC': 'Total Dollar Amount of Fine and Cost of Supervision',
    'AMTREST': 'Dollar Amount of Restitution Ordered',
    'AMTTOTAL': 'Sum of Fine, Cost of Supervision, and Restitution',
    'ALTDUM': 'Indicator of Alternative Sentence',
    'ALTMO': 'Total Months of Alternative Incarceration',
    'TIMSERVD': 'Amount of Time Already Served',
    'SUPRDUM': 'Indicator of Supervised Release',
    'SUPREL': 'Number of Months of Supervised Release Ordered'
}

# Define the category labels for specific variables
category_labels = {
    'MONSEX': {0: 'Male', 1: 'Female'},
    'NEWRACE': {1: 'White', 2: 'Black', 3: 'Hispanic', 6: 'Other'},
    'NEWCIT': {0: 'U.S.', 1: 'Non-U.S. (Includes Legal and Illegal Aliens)'},
    'HISPORIG': {0: 'Info on Hispanic Origin Not Available', 1: 'Non-Hispanic', 2: 'Hispanic'},
    'CRIMHIST': {0: 'No Criminal History', 1: 'Yes, There is Criminal History'},
    'ACCAP': {0: 'Not Applied', 1: 'Applied'},
    'WEAPSOC': {0: 'No SOC Weapon Enhancement Present', 1: 'SOC Weapon Enhancement Present'},
    'DISPOSIT': {0: 'No Imprisonment', 1: 'Guilty Plea', 2: 'Nolo Contendere', 3: 'Jury Trial', 4: 'Trial by Judge or Bench Trial', 5: 'Guilty Plea and Trial (>1 Count)'}
}

# Iterate over each input and output variable
for input_var, input_desc in input_vars.items():
    for output_var, output_desc in output_vars.items():
        # Create a subplot for input vs input count, input vs output, and input vs average output
        fig, axs = plt.subplots(1, 3, figsize=(20, 6))
        
        # Create category labels explanation for the plot title
        category_labels_str = ', '.join([f'{k} = {v}' for k, v in category_labels.get(input_var, {}).items()])
        fig.suptitle(f'{input_desc} vs {output_desc}\n({category_labels_str})')
        
        # Plot input vs output
        if data[input_var].dtype == 'object' and data[output_var].dtype == 'object':
            # Categorical input and output
            pd.crosstab(data[input_var], data[output_var]).plot(kind='scatter', ax=axs[0])
        elif data[input_var].dtype == 'object':
            # Categorical input and numerical output
            data.groupby(input_var)[output_var].mean().plot(kind='scatter', ax=axs[0])
        elif data[output_var].dtype == 'object':
            # Numerical input and categorical output
            for category in data[output_var].unique():
                data[data[output_var] == category].plot(kind='scatter', x=input_var, y=output_var, label=category, ax=axs[0])
            axs[0].legend()
        else:
            # Numerical input and output
            data.plot(kind='scatter', x=input_var, y=output_var, ax=axs[0])
        axs[0].set_xlabel(input_desc)
        axs[0].set_ylabel(output_desc)
        axs[0].set_title('Input vs Output')
        
        # Plot input vs count
        data[input_var].value_counts().plot(kind='bar', ax=axs[1])
        axs[1].set_xlabel(input_desc)
        axs[1].set_ylabel('Count')
        axs[1].set_title('Input vs Count')
        
        # Plot input vs average output
        if data[output_var].dtype != 'object':
            data.groupby(input_var)[output_var].mean().plot(kind='bar', ax=axs[2])
            axs[2].set_xlabel(input_desc)
            axs[2].set_ylabel(f'Average {output_desc}')
            axs[2].set_title('Input vs Average Output')
        else:
            axs[2].axis('off')
        
        plt.tight_layout()
        plt.subplots_adjust(top=0.75)
        
        # Save the plot
        plt.savefig(f'plots/{input_var}_{output_var}.png')
        plt.close()
```
